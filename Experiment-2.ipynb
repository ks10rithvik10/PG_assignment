__Question 2__
📝 Aim

To calculate the area of a circle using the radius entered by the user.

__Explanation__

The area of a circle is calculated using the formula π × r². The program accepts the radius as input from the user and uses the math.pi constant for an accurate value of π.

```python
import math

radius = float(input("Enter the radius: "))

area = math.pi * radius ** 2

print("Area of the circle =", area)
```
